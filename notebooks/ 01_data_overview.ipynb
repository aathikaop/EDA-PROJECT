### Data Quality Issues Identified

| Issue | Column | Description | Planned Action |
|-----|------|------------|----------------|
| Missing values | Family_Income, Study_Hours_per_Day, Stress_Index, Parental_Education | Around 5% missing entries present | Impute using median,mean(numerical) and mode(categorical)|
| Incorrect data type | Age | Stored as float initially but represents integer values | Convert to int |
| Potential outliers | Family_Income, Study_Hours_per_Day, Travel_Time_Minutes | Extreme values may distort analysis | Detect using IQR method|
