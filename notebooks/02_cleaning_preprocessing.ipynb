```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../data/raw/student_dropout_dataset_v3.csv")
```

```python
df
```

```python
df['Age']=df['Age'].round().astype(int)
```

```python
df.dtypes
```

```python
df['Family_Income'].fillna(df['Family_Income'].median(),inplace=True)
```

```python
df['Study_Hours_per_Day'].fillna(df['Study_Hours_per_Day'].mean(), inplace=True)
```

```python
df['Stress_Index'].fillna(df['Stress_Index'].mean(), inplace=True)
```

```python
df['Parental_Education'].fillna(df['Parental_Education'].mode()[0], inplace=True)
```

### Observation
- Family_Income was imputed using median due to strong positive skew.
- Study_Hours_per_Day and Stress_Index were imputed using mean as their distributions were symmetric.
- Parental_Education was imputed using mode to preserve categorical balance.
- No missing values remain after treatment.

```python
outlier_checking_cols = [
    'Family_Income',
    'Study_Hours_per_Day',
    'Travel_Time_Minutes'
]
```

```python
for col in outlier_checking_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col} → {len(outliers)} outliers")
```

### Outlier Detection Observations

- Most detected outliers are distribution-driven rather than data errors.

- Family_Income shows high outlier count due to right-skewed income distribution.

- Other variables show realistic extreme values and were retained.

- No outlier removal performed to preserve real-world variability.

- Outliers were retained since removal could distort behavioral patterns and model performance.
