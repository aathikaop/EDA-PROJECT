## Dropout Distribution

```python
df["Dropout"].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Dropout", data=df)
plt.title("Dropout Distribution")
plt.xlabel("Dropout")
plt.ylabel("Frequency")
plt.savefig("../reports/figures/dropout_distribution.png", bbox_inches="tight")
plt.show()
```

The distribution shows that around 23.5% of students dropped out, while 76.5% continued their studies. This indicates that dropout is a significant issue affecting nearly one in four students, and further analysis is required to identify the key factors contributing to it.

## Correlation Heatmap

```python
num_df = df.select_dtypes(include=["int64", "float64"]).drop(columns=["Student_ID"])
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
plt.savefig('Correlation_Heatmap.png',dpi=500,bbox_inches='tight')
```

- Dropout and GPA have negative correlation.As GPA decreases, the likelihood of dropping out increases.
- Lower attendance correlates slightly with higher dropout rates.
- Stress Index and Dropout have positive correlation.Means high Stress index may effect dropout.
- Higher stress is correlated with lower GPAs.

## Key Insights
- Dropout is strongly influenced by academic performance, particularly GPA
- Higher stress levels are associated with increased dropout risk
- Attendance has a mild influence on dropout outcomes
- Academic performance variables dominate over demographic factors
