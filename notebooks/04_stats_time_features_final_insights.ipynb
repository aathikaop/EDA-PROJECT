```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
```

```python
df = pd.read_csv("../data/interim/cleaned_day2.csv")
```

```python
df.head()
```

## Two Sample Independent T-test (GPA)

Checking whether if there is any mean difference between GPA of dropout students and non-dropout students.

- Null hypothesis(H0):there is no difference between GPA of dropout students and non-dropout students.
- Alternative hypothesis(H1):There is a significant difference between GPA of dropout students and non-dropout students.

```python
dropout_yes = df[df["Dropout"] == 1]["GPA"]
dropout_no = df[df["Dropout"] == 0]["GPA"]

t_stat, p_value = ttest_ind(dropout_yes, dropout_no)

print("T-stat:", t_stat)
print("P-value:", p_value)

if p_value <=0.05:
  print('There is significance difference between the mean GPA between dropout and non dropout.')
else:
  print('There is no difference in GPA mean between dropout and non dropout.')
```

## Interpretation
- Result:The p-value is extremely small (p < 0.05), which means the difference is statistically significant.
- conclusion:GPA has a strong relationship with dropout.

# Two Sample Independent T-test (Stress Index)

Checking whether if there is any mean difference between GPA of dropout students and non-dropout students.

- Null hypothesis(H0): There is no difference between stress level of dropout students and non-dropout students.
- Alternative hypothesis(H1): There is a significant difference between stress level of dropout students and non-dropout students


```python
stress_dropout = df[df["Dropout"] == 1]["Stress_Index"]
stress_no_dropout = df[df["Dropout"] == 0]["Stress_Index"]
t_stat, p_value = ttest_ind(stress_dropout,stress_no_dropout)

print("T-stat:", t_stat)
print("P-value:", p_value)

if p_value <=0.05:
  print('There is significance difference between the mean stress level between dropout and non dropout.')
else:
  print('There is no difference in mean stress level between dropout and non dropout.')
```

## Interpretation
- The p-value is far below 0.05, so we reject the null hypothesis.
- Mean stress level for students who dropped out is significantly higher than for those who didn’t drop.
- Conclusion: Higher stress levels are strongly associated with increased dropout risk.

## Two Sample Independent T-test (Attendence Rate)

Checking whether if there is any mean difference between attendence of dropout students and non-dropout students.

- Null hypothesis(H0): There is no difference between attedence of dropout students and non-dropout students.
- Alternative hypothesis(H1): There is a significant difference between attendence of dropout students and non-dropout students.

```python
dropout = df[df["Dropout"] == 1]["Attendance_Rate"]
no_dropout = df[df["Dropout"] == 0]["Attendance_Rate"]
t_stat, p_value = ttest_ind(dropout,no_dropout)

print("T-stat:", t_stat)
print("P-value:", p_value)

if p_value <=0.05:
  print('There is significance difference between the mean attendence between dropout and non dropout.')
else:
  print('There is no difference in mean attendence between dropout and non dropout.')
```

## Interpretation

- The p-value is below 0.05, so we reject the null hypothesis.
- Attendance rate for students who dropped out is significantly lower than for those who didn’t drop.

## ANOVA Test (GPA vs Department)

To check if the mean GPA differs across departments.

```python
groups = [group['GPA'].values for name, group in df.groupby('Department')]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a statistically significant difference in mean GPA across departments.")
else:
    print("There is no statistically significant difference in mean GPA across departments.")
```

## Interpretation
- The p-value is greater than 0.05.
- we fail to reject the null hypothesis.
- Conclusion: there is no statistically significant difference in mean GPA among the departments.

# Feature Engineering

## Study Efficiency
- Study efficiency is calculated for how much effort a student puts in for each unit of their GPA.
- Higher values might indicate less efficiency or greater difficulty.
-  this tells how many study hours are required to produce one single point of GPA.

```python
df["Study_Efficiency"] = df["Study_Hours_per_Day"] / df["GPA"]
df
```

## Academic Habits Score

- Measures student engagement and compliance by dividing the Attendance_Rate by the Assignment_Delay_Days plus 1.
- A higher score indicates better attendance and fewer delays.

```python
df["Academic_Habits_Score"]=df['Attendance_Rate'] / (df['Assignment_Delay_Days'] + 1)
df
```

## CGPA Improvement

- This measures the difference between a student's current semester performance and their cumulative performance over time.

### Interpretation:

-Positive Value: The student's grades are improving.

-Negative Value: The student's grades are declining.

```python
df['CGPA_Improvement']=df['Semester_GPA']-df['CGPA']
df
```
